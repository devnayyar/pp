```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import random as rd
```

```python
ds_heart = pd.read_csv("heart.csv")
```

```python
ds_heart.info()
```

```python
ds_heart.head()
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
plt.scatter(x='Age',y='RestBP', data=ds_heart, c='g') 
plt.scatter(x='Age',y='Chol', data=ds_heart, c='r', marker='+') 
plt.xlabel('Age')
plt.ylabel('Chol and RestBP') 
plt.legend(title='scatter_plot') 
plt.show()
```

```python
plt.figure() 
plt.legend(title='Scatter_plot')
sns.lmplot(x='Age', y='Chol', data=ds_heart, fit_reg=False)
```

```python
plt.figure() 
plt.plot(ds_heart['Chol']) 
plt.plot(ds_heart['RestBP']) 
plt.legend(title=' Line_graph') 
plt.show()
```

```python
plt.figure() 
plt.legend(title='line_graph')
sns.lineplot(x="Age", y="Chol", data=ds_heart)

```

```python
plt.figure() 
plt.xlabel('ChestPain') 
plt.ylabel('Count') 
plt.legend(title='Bar graph')
ds_heart.ChestPain.value_counts().plot(kind='bar')
```

```python
plt.figure() 
plt.legend(title='Bar_graph') 
sns.countplot(ds_heart.ChestPain)
```

```python
plt.figure() 
plt.legend(title='Histogram') 
ds_heart.plot(kind='hist')
```
