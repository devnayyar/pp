```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import random as rd
```

```python
ds_heart = pd.read_csv("heart.csv")
```

```python
ds_heart.info()
```

```python
ds_heart.head()
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
plt.figure() 
plt.legend(title='Scatter_plot')
sns.lmplot(x='Age', y='Chol', data=ds_heart, fit_reg=False)
```

```python
plt.figure() 
plt.legend(title='line_graph')
sns.lineplot(x="Age", y="Chol", data=ds_heart)
```

```python
plt.figure() 
plt.legend(title='Bar_graph') 
sns.countplot(ds_heart.ChestPain)
```

```python
plt.figure()
sns.displot(ds_heart.Age)
```

```python
factors = ['Age','Chol','RestBP','RestECG','Slope']
data=[0.4,0.8,1,1.5,0.9]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=factors)
plt.show()
```

```python
data=[0.4,0.8,1,1.5,0.9]
labels=['Age','Chol','RestBP','RestECG','Slope']
colors=sns.color_palette('pastel')[0:5]
plt.pie(data,labels=labels,colors=colors,autopct='%.0f%%')
plt.show()
```
