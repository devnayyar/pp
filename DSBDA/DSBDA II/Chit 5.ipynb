```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
ds_heart = pd.read_csv("heart.csv")
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
plt.scatter(x='Age',y='RestBP', data=ds_heart, c='g')
plt.scatter(x='Age',y='Chol', data=ds_heart, c='r', marker='+')
plt.xlabel('Age')
plt.ylabel('Chol and RestBP')
plt.legend(title='scatter_plot')
plt.show()
```

```python
plt.figure()
plt.xlabel('ChestPain')
plt.ylabel('Count')
plt.legend()
ds_heart.ChestPain.value_counts().plot(kind='bar')
```

```python
plt.figure()
plt.plot(ds_heart['Chol'])
plt.plot(ds_heart['RestBP'])
plt.show()
```

```python
plt.figure()
sns.lineplot(x="Age", y="Chol", data=ds_heart)
```

```python
factors = ['Age', 'Chol', 'RestBP', 'Slope']
data=[0.4,0.8,1,1.5]
fig=plt.figure(figsize=(4,4))
plt.pie(data,labels=factors)
plt.show()
```

```python
plt.figure()
sns.lmplot(x='Age', y='Chol', data=ds_heart, fit_reg=False)
```
